### Train Test Split 

```python
from sklearn.model_selection import train_test_split
```

```python
y=df.pop('cnt')
X=df
```

```python
X_train, X_test, y_train, y_test= train_test_split(X,y, test_size=0.30, random_state=42)
```

### Feature Scaling 

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
col=[f for f in X_train.columns]
col
```

```python
scaler = MinMaxScaler()
```

```python
X_train[col]=scaler.fit_transform(X_train[col])
```

```python
X_train
```

```python
plt.figure(figsize=(27,12))
sns.heatmap(X_train.corr(), annot=True, cmap='YlGnBu')
```

### Feature Selection Using RFE 

```python
from sklearn.feature_selection import RFE
from sklearn.linear_model import LinearRegression
```

```python
reg = LinearRegression()
reg.fit(X_train, y_train)
rfe = RFE(reg, n_features_to_select=10)
rfe = rfe.fit(X_train, y_train)
```

```python
list(zip(X_train.columns, rfe.support_, rfe.ranking_))
```

```python
l1=X_train.columns[rfe.support_]
l1
```

### Building a Model Using StatsModel

```python
#!pip install statsmodels==0.13.5
```

```python
import statsmodels.api as sm
```

```python
X_train_sm= X_train[l1]
X_train_sm= sm.add_constant(X_train_sm)
X_train_sm
```

```python
lr= sm.OLS(y_train, X_train_sm)
lr_model=lr.fit()
lr_model.summary()
```

### Calculate VIF

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
```

```python
vif = pd.DataFrame()
y=X_train_sm.pop('const')
X= X_train_sm
vif['Features']=X.columns
vif['VIF']=[variance_inflation_factor(X.values, i) for i in range(X.shape[1])]
vif['VIF']= round(vif['VIF'],2)
vif = vif.sort_values(by = 'VIF', ascending=False)
vif
```

```python
X_train_sm=X_train[l1]
```

```python
# hum vif value is 17.22 so this shows multicolinearity with other independent variables
X_train_sm.drop('hum', axis=1, inplace=True)
```

```python
X_train_sm=sm.add_constant(X_train_sm)
```

```python
# Rebuilding a model
lr= sm.OLS(y_train, X_train_sm)
lr_model=lr.fit()
lr_model.summary()
```

```python
vif = pd.DataFrame()
y=X_train_sm.pop('const')
X= X_train_sm
vif['Features']=X.columns
vif['VIF']=[variance_inflation_factor(X.values, i) for i in range(X.shape[1])]
vif['VIF']= round(vif['VIF'],2)
vif = vif.sort_values(by = 'VIF', ascending=False)
vif
```

```python
X_train_sm=sm.add_constant(X_train_sm)
```

```python
y_train_pred=lr_model.predict(X_train_sm)
```
